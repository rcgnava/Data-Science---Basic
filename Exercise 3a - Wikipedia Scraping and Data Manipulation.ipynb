```python
import pandas as pd
import numpy as np
import requests
import urllib.request

url = 'https://en.wikipedia.org/wiki/40_Greatest_Players_in_PBA_History'
page = urllib.request.urlopen(url)

from bs4 import BeautifulSoup
soup = BeautifulSoup(page, 'html.parser')

#1: Determine data to extract
text = soup.get_text()
all_tables=soup.find_all('table')
right_table=soup.find('table', class_='wikitable sortable')

#2: Scraping rows and headers, cleaning \n values in the data 
headers = []
headers = [header.text for header in right_table.find_all('th')]
headers = [i.replace("\n", "") for i in headers]

rows = []
for row in right_table.find_all('tr'):
    cells = row.find_all('td')
    if len(cells) > 0:
        name_link = cells[0].find('a')
        name_info = [cell.text.strip('\n') for cell in cells]
        rows.append(name_info)
        
df1 = pd.DataFrame(rows, columns=headers)

#4: Replacing NAN values with another value
df1.fillna("none", inplace=True)

#5: Dropping irrelevant columns since they are all none values
df1.drop('Ref', axis='columns', inplace=True)
df1.drop('All-Star', axis='columns', inplace=True)
df1
```
