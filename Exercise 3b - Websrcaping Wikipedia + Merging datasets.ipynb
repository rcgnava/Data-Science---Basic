```python
import pandas as pd
import numpy as np
import requests

#Parsing data 
website_url = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
from bs4 import BeautifulSoup
soup = BeautifulSoup(website_url, 'html.parser')

## DATA 1
#1. Finding specific data table in Wikipedia
my_table = soup.find('table', {'class': 'wikitable sortable'})

#2. Scraping data - header
headers = []
headers = [header.text for header in my_table.findAll('th')]
headers = [i.replace("\n", "") for i in headers]

#3. Scraping data - rows
rows = []
for row in my_table.find_all('tr'):
    cells = row.find_all('td')
    if len(cells) > 0:
        name_link = cells[0].find('a')
        name_info = [cell.text.strip('\n') for cell in cells]
        rows.append(name_info)
        
df1 = pd.DataFrame(rows, columns=headers)

## DATA 2
# 1. Import data from CSV file
df2 = pd.read_csv("Datasets/Geospatial_Coordinates.csv", index_col="Postcode")

## Merging 2 datasets and filling NaN values with a value 
df3 = pd.merge(df1,df2, how='left', on='Postcode')
df3.fillna('0.00',  inplace=True)
df3
```
